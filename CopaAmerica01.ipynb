```python
def extract_first_name(full_name):
    # Split the full name and return the first part
    return full_name.split()[0] if isinstance(full_name, str) else None

# Fill None values in 'player_nickname' with the first name from 'player_name'
Argentina_Lineup['player_nickname'] = Argentina_Lineup.apply(
    lambda row: extract_first_name(row['player_name']) if pd.isnull(row['player_nickname']) else row['player_nickname'],
    axis=1
)

# Print the DataFrame to check the results
Argentina_Lineup
```

```python


# "change make the horizantal dictionary into vertical dataframe'

# Define the function to extract position data from the DataFrame
def extract_position_data(df):
    results = []
    for index, player in df.iterrows():
        positions = player['positions']
        if isinstance(positions, list) and len(positions) > 0:
            # Assuming we are interested in the first position entry
            position = positions[0]
            result = {
                'Player': player['player_nickname'],
                'Position ID': position.get('position_id'),
                'Position': position.get('position'),
                'From': position.get('from'),
                'To': position.get('to'),
                'From Period': position.get('from_period'),
                'To Period': position.get('to_period'),
                'Start Reason': position.get('start_reason'),
                'End Reason': position.get('end_reason')
            }
            results.append(result)
    return pd.DataFrame(results)

# Store the function in a variable
position_data_extractor = extract_position_data

# Call the function with your DataFrame
Argentina_Lineup_df = position_data_extractor(Argentina_Lineup)

# Display the DataFrame with the extracted position data
print(Argentina_Lineup_df)
```

```python
Argentina_Starting_XI = Argentina_Lineup_df[Argentina_Lineup_df['Start Reason'] == 'Starting XI']
Argentina_Starting_XI
```

```python
Argentina_Substitutions = Argentina_Lineup_df[Argentina_Lineup_df['Start Reason'] != 'Starting XI']
Argentina_Substitutions
```

```python
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
import pandas as pd

# Define a mapping from position names to pitch coordinates
position_to_coords = {
    'Goalkeeper': (1, 40),
    'Right Back': (18, 10),
    'Right Center Back': (18, 30),
    'Center Back': (18, 40),
    'Left Center Back': (18, 50),
    'Left Back': (18, 70),
    'Right Wing': (55, 10),
    'Right Midfield': (40, 20),
    'Center Midfield': (35, 40),
    'Left Midfield': (40, 55),
    'Left Wing': (55, 70),
    'Right Forward': (70, 25),
    'Center Forward': (55, 40),
    'Left Forward': (70, 55),
    'Right Wing Back': (35, 10),
    'Left Wing Back': (35, 70),
    'Left Defensive Midfield': (30, 50),
    'Right Defensive Midfield': (30, 30),
    'Center Attacking Midfield': (45, 40),
    'Right Center Forward': (55, 30),  # Added position
    'Left Center Forward': (55, 50)    # Added position
}

# Example DataFrame (replace with your actual DataFrame)
# Argentina_Starting_XI = pd.DataFrame({
#     'player_name': ['Lionel Messi', 'Ángel Di María', 'Lautaro Martínez'],
#     'position': ['Right Wing', 'Left Wing', 'Center Forward']
# })

# Extract coordinates for the players
players_coordinates = []
for index, row in Argentina_Starting_XI.iterrows():
    position = row['Position']
    player_name = row['Player']
    if position in position_to_coords:
        coord = position_to_coords[position]
        players_coordinates.append((player_name, coord))
    else:
        print(f"Position {position} not found in mapping for player {player_name}")

# Check the number of players in the starting XI
print(f"Total players in starting XI: {len(players_coordinates)}")
for player_name, coord in players_coordinates:
    print(f"Player: {player_name}, Coordinates: {coord}")

# Plot the pitch
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='grass',stripe_color='#c2d59d', line_color='white',stripe=True)
fig, ax = pitch.draw(figsize=(7, 10))

# Plot the players
for player_name, coord in players_coordinates:
    x, y = coord
    pitch.scatter(x, y, ax=ax, color='#43A1D5', s=300, edgecolors='black')
    pitch.annotate(player_name, xy=(x, y), ax=ax, ha='center', va='center', color='black', fontsize=10, weight='bold')

plt.show()
```
