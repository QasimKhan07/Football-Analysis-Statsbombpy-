```python
def extract_first_name(full_name):
    # Split the full name and return the first part
    return full_name.split()[0] if isinstance(full_name, str) else None

# Fill None values in 'player_nickname' with the first name from 'player_name'
Argentina_Lineup['player_nickname'] = Argentina_Lineup.apply(
    lambda row: extract_first_name(row['player_name']) if pd.isnull(row['player_nickname']) else row['player_nickname'],
    axis=1
)

# Print the DataFrame to check the results
Argentina_Lineup
```

```python
# player_positions = []
for index, player in Argentina_Lineup.iterrows():
    positions = player['positions']
    if isinstance(positions, list) and len(positions) > 0:
        position = positions[0]
        print(f"Player: {player['player_name']}, Position Data: {position}")  # Print the position data to inspect its structure
```

```python


# "change make the horizantal dictionary into vertical dataframe'

# Define the function to extract position data from the DataFrame
def extract_position_data(df):
    results = []
    for index, player in df.iterrows():
        positions = player['positions']
        if isinstance(positions, list) and len(positions) > 0:
            # Assuming we are interested in the first position entry
            position = positions[0]
            result = {
                'Player': player['player_nickname'],
                'Position ID': position.get('position_id'),
                'Position': position.get('position'),
                'From': position.get('from'),
                'To': position.get('to'),
                'From Period': position.get('from_period'),
                'To Period': position.get('to_period'),
                'Start Reason': position.get('start_reason'),
                'End Reason': position.get('end_reason')
            }
            results.append(result)
    return pd.DataFrame(results)

# Store the function in a variable
position_data_extractor = extract_position_data

# Call the function with your DataFrame
Argentina_Lineup_df = position_data_extractor(Argentina_Lineup)

# Display the DataFrame with the extracted position data
print(Argentina_Lineup_df)
```

```python
# Filter events for Argentina players
argentina_pass_receptions = events[(events['team'] == 'Argentina') & (events['type'] == 'Pass')]

# Extract necessary columns
receptions = argentina_pass_receptions[['player', 'location', 'pass_recipient', 'timestamp']]
receptions
```

```python
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
import pandas as pd
import numpy as np



# Simulating multiple pass receptions to create more location points for the heatmaps
additional_data = []  # Store additional rows

for i in range(10):  # Adding random locations to simulate more activity (replace with real data)
    additional_locations = [[np.random.randint(0, 120), np.random.randint(0, 80)] for _ in range(11)]
    for j, loc in enumerate(additional_locations):
        additional_data.append({'player': receptions['player'][j], 'location': loc, 'pass_recipient': receptions['pass_recipient'][j]})

# Convert additional data into a DataFrame
additional_receptions = pd.DataFrame(additional_data)

# Concatenate the original receptions DataFrame with the additional data
receptions = pd.concat([receptions, additional_receptions], ignore_index=True)

# Define a vertical pitch with green grass color
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='white', line_color='black')

# Create the figure for the overall formation with a larger size
fig, ax = plt.subplots(figsize=(12, 16))  # Increased figure size for a larger pitch
pitch.draw(ax=ax)  # Draw the main pitch

# Define the relative positions for each player in a 4-3-3 formation
positions = {
    'Emiliano Martinez': [51, 20],   # Goalkeeper
    'Nicolas Tagliafico': [25, 35],  # Left Back
    'Cristian Romero': [40, 35],  # Center Back 1
    'Lisandro Martinez': [60, 35],  # Center Back 2
    'Nahuel Molina': [75, 35],  # Right Back
    'Rodrigo De Paul': [35, 55],  # Central Midfielder 1 (Left)
    'Enzo Fernandez': [50, 50],  # Central Midfielder 2 (Center)
    'Alexis Mac Allister': [65, 55],  # Central Midfielder 3 (Right)
    'Nicolas Gonzalez': [25, 65],  # Left Forward
    'Julian Alvarez': [50, 75],  # Central Forward
    'Lionel Messi': [75, 65]   # Right Forward
}

# Plot subplots for each player's heatmap
for player, (x, y) in positions.items():
    ax_sub = fig.add_axes([x / 100 - 0.05, y / 100 - 0.05, 0.1, 0.1], zorder=2)  # Position of each subplot
    pitch.draw(ax=ax_sub)  # Draw individual pitch for each player

    # Extract player data
    player_data = receptions[receptions['pass_recipient'] == player]
    if not player_data.empty:
        locations = list(player_data['location'])
        x_locs, y_locs = zip(*locations)
        pitch.kdeplot(x_locs, y_locs, ax=ax_sub, shade=True, levels=50, shade_lowest=False, alpha=0.5)

    # Add player name from the starting 11 DataFrame
    ax_sub.text(60, 85, player, fontsize=10, ha='center', va='center', color='black', fontweight='bold')

# Set title and display plot
plt.suptitle('Pass Receptions - Argentina vs. Colombia (2024 Copa Final) - 4-3-3 Formation', fontsize=20, y=0.9)
plt.show()
```
